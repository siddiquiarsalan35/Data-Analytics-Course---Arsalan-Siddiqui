```python
## Question 05

text =  "Hello Python World"
print (text)

modified_text = text.replace (" ", "_")  ## Replacing spaces with underscores
print (modified_text)

words_list = modified_text.split("_")  ## Split the modified string into the list
print (words_list)

final_text = "-".join(words_list)  ## Joins the list with hyphens
print (final_text)
```

```python
## Question 09

text = "Data Analytics with Python is very Powerfull"

print (text)

clean_text = text.strip()  ## Removing extra spaces
print (clean_text)

title_text = clean_text.title()  ## converting to title case
print (title_text)

updated_text = title_text.replace("Powerfull", "Important")  ##updating text
print (updated_text)

python_position = updated_text.find("Python")  ##Finding position of python
print (python_position)
```
